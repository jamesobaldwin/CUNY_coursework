```python
# Catalan numbers! From example 2.1 

def Catalan(N):
    """prints the first N Catalan numbers"""
    n=0       # initialize n=0
    Cn=1     # initialize C0= 1
    while(n<N):
        print('n=', n, 'Cn=', int(Cn))
        Cn = Cn * (4*n+2)/(n+2)
        n=n+1

def Catalan_up_to_value(value):
    """prints all Catalan numbers below a specified value"""
    n=0       # initialize n=0
    Cn=1     # initialize C0= 1
    while(Cn<value):
        print('n=', n, 'Cn=', int(Cn))
        Cn = Cn * (4*n+2)/(n+2)
        n=n+1

# Catalan(15)

Catalan_up_to_value(10000)
```

```python
import numpy as np 

def Madelung(L): 
    M=0
    for x in range(-1*L, L+1):
        for y in range(-1*L, L+1):
            for z in range(-1*L, L+1):            
                if ((x==0) and (y==0) and z==0): 
                    M+=0 
                else: 
                    p1 = (-1)**(x+y+z)/np.sqrt(x**2+y**2+z**2) # potential from one
                    M += p1 
    return np.abs(M)
print('10', Madelung(10))    
print('50',Madelung(50))    
print('100',Madelung(100))    
```

```python
def Madelung(L):
    M = 0
    
    for a in range(-1*L, L+1): 
        for b in range(-1*L, L+1): 
            for c in range(-1*L, L+1): 
                
            
                if ((a==0) and (b==0) and (c==0)): 
                    M+=0
                else: 
                    M+= (-1)**(a+b+c)/np.sqrt(a**2+b**2+c**2)
    return M
print('truncating at \pm 10 neighbors: ', Madelung(10))
print('truncating at \pm 100 neighbors: ', Madelung(100))
```

```python
# Madelung function Example 2.9

def Madelung(L=100):
    
    M = 0
    for i in range(-L, L+1):
        for j in range(-L, L+1):
            for k in range(-L, L+1):

                if (i==j==k==0):
                    continue

                M += (-1)**(i+j+k)/((i**2+j**2+k**2)**(1/2))

    print(f"The Madelung constant of NaCl is {abs(round(M,4))}")   
    return M 

print(Madelung())
```

```python
def MadelungV(L=100, a=1):
    Const = 1.112e-10 # C**2 J**-1 m**-1 (represents 4*pi*eps0)
    e = 1.6022e-19 # electric charge in Coulombs
    
    M = 0
    for i in range(-L, L+1):
        for j in range(-L, L+1):
            for k in range(-L, L+1):

                if (i==j==k==0):
                    continue

                M += (-1)**(i+j+k)/((i**2+j**2+k**2)**(1/2))
    V = e/(Const*a) * M
    print(f"The Madelung constant of NaCl is {abs(round(M,4))}")   
    print(f"The total voltage of NaCl given a is ", V)       
    return V


print(MadelungV())
```
