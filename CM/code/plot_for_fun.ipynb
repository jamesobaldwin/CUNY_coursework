```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes, mark_inset

# Create the figure and axis
fig, ax = plt.subplots(figsize=(8, 8))

# Define the Deltoid curve (scaled down)
t = np.linspace(0, 2 * np.pi, 1000)
x_deltoid = (2 * np.cos(t) + np.cos(2 * t)) * 0.008  # Further scaling down
y_deltoid = (2 * np.sin(t) - np.sin(2 * t)) * 0.008

# Define Fey's function (scaled down)
theta_fey = np.linspace(0, 24 * np.pi, 1000)
r_fey = np.exp(np.sin(theta_fey)) - 2 * np.cos(4 * theta_fey) + np.sin((2 * theta_fey - np.pi) / 24) ** 5
x_fey = r_fey * np.cos(theta_fey) * 0.007  # Further scaling down
y_fey = r_fey * np.sin(theta_fey) * 0.007

# Define Galilean Spiral (much larger scale)
theta_spiral = np.linspace(0, 10 * np.pi, 1000)
r_spiral = theta_spiral ** 2
x_spiral = r_spiral * np.cos(theta_spiral)
y_spiral = r_spiral * np.sin(theta_spiral)

# Plot the Galilean spiral
ax.plot(x_spiral, y_spiral, label="Galilean Spiral", color="blue", alpha=0.5)

# Plot the Deltoid curve (scaled down)
ax.plot(x_deltoid, y_deltoid, label="Deltoid Curve", color="red")

# Plot Fey's function (scaled down)
ax.plot(x_fey, y_fey, label="Fey's Function", color="green")

# Zoomed-in box for Fey's function (upper left inset)
axins1 = zoomed_inset_axes(ax, zoom=10000, loc='upper left')  # Increased zoom level to 10000
axins1.plot(x_fey, y_fey, color="green")
axins1.set_xlim([-0.03, 0.03])  # Tighter limits to show the tiny curves
axins1.set_ylim([-0.03, 0.03])
axins1.set_xticks([])
axins1.set_yticks([])

# Zoomed-in box for Deltoid curve (lower right inset)
axins2 = zoomed_inset_axes(ax, zoom=10000, loc='lower right')  # Increased zoom level to 10000
axins2.plot(x_deltoid, y_deltoid, color="red")
axins2.set_xlim([-0.04, 0.04])  # Tighter limits to show the tiny curves
axins2.set_ylim([-0.02, 0.02])
axins2.set_xticks([])
axins2.set_yticks([])

# Connecting the upper left zoomed-in box with the main plot
mark_inset(ax, axins1, loc1=2, loc2=1, fc="none", ec="0.5")  # Add lines for Fey's function
mark_inset(ax, axins1, loc1=3, loc2=4, fc="none", ec="0.5")  # Add lines for Fey's function

# Connecting the lower right zoomed-in box with the main plot
mark_inset(ax, axins2, loc1=2, loc2=4, fc="none", ec="0.5")  # Already in the code
mark_inset(ax, axins2, loc1=1, loc2=3, fc="none", ec="0.5")  # Already in the code


# Set plot title and legend
ax.set_title("Deltoid Curve, Fey's Function, and Galilean Spiral (Scaled View)", fontsize=14)
ax.legend()

plt.savefig("perfect_plot.png")
# Show the plot
plt.show()
```
