```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.linalg import eigh, eigvalsh, solve, inv
from scipy.linalg import solve_banded
from scipy.integrate import quad, simpson
```

```python
# Constants
hbar = 1.0545718e-34  # Js
M = 9.1094e-31        # kg (mass of electron)
L = 5e-10             # meters (width of well)
a = 10 * 1.6022e-19   # Convert eV to Joules
J = 1.602e-19         # for Joule <-> eV conversion

# Function for potential V(x) = ax/L
def V(x):
    return (a * x) / L

# Define the wavefunction psi_n(x)
def psi_n(n, x):
    return np.sin(n * np.pi * x / L)

# Function to compute H_mn based on the given rules
def H_mn(m, n):
    if m == n:
        # Diagonal term when m == n
        return (hbar*np.pi*n)**2/(2*M*L**2) + (a/2)
    elif m % 2 != n % 2:
        # If one is even and other is odd
        return -((8*a) / np.pi**2) * (m * n) / (m**2 - n**2)**2
    else:
        # If both even or both odd
        return 0

#Set up the matrix H for m, n up to 10
size = 100
H = np.zeros((size, size))

# Fill the matrix
for m in range(1, size+1):
    for n in range(1, size+1):
        H[m-1, n-1] = H_mn(m, n)

# # Print the resulting matrix H
# # print("Matrix H:")
# # print(H)

# # calculate and show the eigenvalues
# eigens_100 = eigvalsh(H)/J
# print("Energy levels of infinite well (first 10):\n")
# for eigenvalue in eigens_100:
#     print(f"{eigenvalue:.2f} eV\n")

# Calculate the eigenvalues and eigenvectors
eigenvalues, eigenvectors = eigh(H)

# Normalize the eigenvectors
x_vals = np.linspace(0, L, 1000)
dx = x_vals[1] - x_vals[0]
eigenvectors = eigenvectors / np.sqrt(simpson(eigenvectors**2, dx=dx))

# Function to sum the sine terms weighted by eigenvectors (Fourier coefficients)
def psi(x, eigenvector):
    return sum(coeff * np.sin(n * np.pi * x / L) for n, coeff in enumerate(eigenvector, start=1))

# Calculate the wavefunctions psi(x) for the ground state and first two excited states
psi_ground = psi(x_vals, eigenvectors[:, 0])  # Ground state (first eigenvector)
psi_excited_1 = psi(x_vals, eigenvectors[:, 1])  # First excited state
psi_excited_2 = psi(x_vals, eigenvectors[:, 2])  # Second excited state

# Calculate the probability densities |psi(x)|^2
prob_density_ground = np.abs(psi_ground)**2
prob_density_excited_1 = np.abs(psi_excited_1)**2
prob_density_excited_2 = np.abs(psi_excited_2)**2

# Normalize the wavefunctions (ensuring integral is 1)
norm_ground = simpson(prob_density_ground, x_vals)
norm_excited_1 = simpson(prob_density_excited_1, x_vals)
norm_excited_2 = simpson(prob_density_excited_2, x_vals)

# Normalize the probability densities
prob_density_ground /= norm_ground
prob_density_excited_1 /= norm_excited_1
prob_density_excited_2 /= norm_excited_2

# Plot the probability densities
plt.plot(x_vals, prob_density_ground, label='Ground state (n=1)')
plt.plot(x_vals, prob_density_excited_1, label='First excited state (n=2)')
plt.plot(x_vals, prob_density_excited_2, label='Second excited state (n=3)')

# Add plot labels and title
plt.xlabel('x (meters)')
plt.ylabel('Probability Density |(x)|')
plt.title('Probability Densities for Ground and First Two Excited States')
plt.legend()

# Display the plot
plt.show()
# Check normalization: the integral of |psi(x)|^2 should be approximately 1
print(f"Normalization of Ground State: {norm_ground}")
print(f"Normalization of First Excited State: {norm_excited_1}")
print(f"Normalization of Second Excited State: {norm_excited_2}")
```
